# Michaelis-Menten

The functional form is $$f(x, k_1, k_2) = \frac{k_1 x}{k_2 + x}$$.

It is has linear portion and a non-linear portion.

$$
\frac{\partial f(x, k_1, k_2)}{\partial k_2} = - \frac{k_1 x}{(k_2 + x)^2}
$$

```python
def mm_func(k2_vals, x = None, k1 = None):
    if x is None:
        x = np.array(range(1, 11))
    if k1 is None:
        k1 = 1
    #
    result_arr = np.array([(k1*x)/(k2 + x) for k2 in k2_vals]).T
    df = pd.DataFrame(result_arr, columns=[str(k) for k in k2_vals])
    return df

# Tests
df = mm_func([0.1, 0.5, 1, 2, 3, 10])
x = df.plot()
        
```

It looks like $k_2 = 1$ has a good combination of linearity for small $x$ and flattening for large $x$.
